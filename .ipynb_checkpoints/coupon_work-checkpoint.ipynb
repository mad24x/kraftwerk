### Will a Customer Accept the Coupon?

**Context**

Imagine driving through town and a coupon is delivered to your cell phone for a restaraunt near where you are driving. Would you accept that coupon and take a short detour to the restaraunt? Would you accept the coupon but use it on a sunbsequent trip? Would you ignore the coupon entirely? What if the coupon was for a bar instead of a restaraunt? What about a coffee house? Would you accept a bar coupon with a minor passenger in the car? What about if it was just you and your partner in the car? Would weather impact the rate of acceptance? What about the time of day?

Obviously, proximity to the business is a factor on whether the coupon is delivered to the driver or not, but what are the factors that determine whether a driver accepts the coupon once it is delivered to them? How would you determine whether a driver is likely to accept a coupon?

**Overview**

The goal of this project is to use what you know about visualizations and probability distributions to distinguish between customers who accepted a driving coupon versus those that did not.

**Data**

This data comes to us from the UCI Machine Learning repository and was collected via a survey on Amazon Mechanical Turk. The survey describes different driving scenarios including the destination, current time, weather, passenger, etc., and then ask the person whether he will accept the coupon if he is the driver. Answers that the user will drive there ‘right away’ or ‘later before the coupon expires’ are labeled as ‘Y = 1’ and answers ‘no, I do not want the coupon’ are labeled as ‘Y = 0’.  There are five different types of coupons -- less expensive restaurants (under \\$20), coffee houses, carry out & take away, bar, and more expensive restaurants (\\$20 - \\$50). 

**Deliverables**

Your final product should be a brief report that highlights the differences between customers who did and did not accept the coupons.  To explore the data you will utilize your knowledge of plotting, statistical summaries, and visualization using Python. You will publish your findings in a public facing github repository as your first portfolio piece.

**Investigating the Bar Coupons**

Now, we will lead you through an exploration of just the bar related coupons.  

1. Create a new `DataFrame` that contains just the bar coupons.


```python
#Creating a new data frame for Bar coupons and reseting the index

data_bar = data.query('coupon == "Bar"')
data_bar = data_bar.reset_index(drop=True)

data_bar.head(5)
```

2. What proportion of bar coupons were accepted?

Roughly .59 were accepted


```python
bar_coupons_accepted = data_bar['Coupon Acceptance'].value_counts(normalize = True)

print (bar_coupons_accepted)
```

3. Compare the acceptance rate between those who went to a bar 3 or fewer times a month to those who went more.


```python
lt4_bar_visits = data_bar.query("Bar in ['never', 'less1', '1~3']")["Coupon Acceptance"].value_counts(normalize=True)
gt3_bar_visits = data_bar.query("Bar not in ['never', 'less1', '1~3']")["Coupon Acceptance"].value_counts(normalize=True)

print("visits bar <= 3\n", lt4_bar_visits)
print("visits bar > 3\n", gt3_bar_visits)
```

4. Compare the acceptance rate between drivers who go to a bar more than once a month and are over the age of 25 to the all others.  Is there a difference?


```python
#fixing the data to convert to numeric
data_bar['age'] = data['age'].str.replace('50plus',"51").str.replace("below21","20")

data_bar["age"] = pd.to_numeric(data_bar["age"])

#variable to group all answers categories that show 1+ visits to a bar
bar_ans_mt1 = ["1~3", "4~8", "gt8"]


#finding people over the age of 25 and went to the bar more than 1+
gt25_bar_mt_once = data_bar.query('age > 25 and Bar in @bar_ans_mt1')["Coupon Acceptance"].value_counts(normalize=True)

#finding the remaing people who do not meet the critera above
notgt25_bar_mt_once = data_bar.query('Bar not in ["1~3", "4~8", "gt8", "Unknown"]')["Coupon Acceptance"].value_counts(normalize=True)

print("Greater than 25 and goes to bar more than once a month\n", gt25_bar_mt_once)
print("Remaining people\n", notgt25_bar_mt_once)
```

5. Use the same process to compare the acceptance rate between drivers who go to bars more than once a month and had passengers that were not a kid and had occupations other than farming, fishing, or forestry. 


```python
# Create the query conditions in a variable

query_condition = (
    (data_bar['Bar'].isin(bar_ans_mt1)) &
    (data_bar['passanger'].isin(["Friend(s)", "Partner"])) &
    (data_bar['occupation'] != "Farming Fishing & Forestry")
)

# Creating a new data fram using the Filter based on the query conditions
filtered_data = data_bar[query_condition]

# Calculate normalized value counts for "Coupon Acceptance"
filtered_data["Coupon Acceptance"].value_counts(normalize=True)

```

6. Compare the acceptance rates between those drivers who:

- go to bars more than once a month, had passengers that were not a kid, and were not widowed *OR*
- go to bars more than once a month and are under the age of 30 *OR*
- go to cheap restaurants more than 4 times a month and income is less than 50K. 



```python
# Grouping all answer that are less than 50K
income_lt50 = ["Less than $12500", "$12500 - $24999"  ,"$25000 - $37499","$37500 - $49999"  ]


filter_condition = (
    (data_bar['Bar'].isin(bar_ans_mt1) & (data_bar['passanger'] != "Kid(s)") & (data_bar['maritalStatus'] != "Widowed")) |
    (data_bar['Bar'].isin(bar_ans_mt1) & (data_bar['age'] < 30)) |
    (data_bar['RestaurantLessThan20'].isin(["4~8", "gt8"]) & data_bar['income'].isin(income_lt50))
)

# Calculate value counts for those meeting the criteria
coupon_acceptance = data_bar[filter_condition]["Coupon Acceptance"].value_counts(normalize=True).reset_index()
coupon_acceptance.columns = ['Coupon Acceptance', 'Proportion']
coupon_acceptance['Group'] = 'Criteria Met'

# Calculate value counts for those not meeting the criteria
other_coupon_acceptance = data_bar[~filter_condition]["Coupon Acceptance"].value_counts(normalize=True).reset_index()
other_coupon_acceptance.columns = ['Coupon Acceptance', 'Proportion']
other_coupon_acceptance['Group'] = 'Criteria Not Met'

# Combine both DataFrames for easier plotting
combined_df = pd.concat([coupon_acceptance, other_coupon_acceptance])

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(data=combined_df, x='Coupon Acceptance', y='Proportion', hue='Group')

# Customize the plot
plt.xlabel('Coupon Acceptance')
plt.ylabel('Proportion')
plt.title('Coupon Acceptance Proportion Comparison')
plt.legend(title='Group')

# Show the plot
plt.show()


```

Additionally, those who often visit bars and are over the age of 25 show a higher likelihood of accepting a coupon. Age appears to be a significant factor, as individuals under 21, who are not of legal drinking age, are less likely to accept a coupon.

### Independent Investigation

Using the bar coupon example as motivation, you are to explore one of the other coupon groups and try to determine the characteristics of passengers who accept the coupons.  

I have chosen to look at the "Coffee House" coupons

Create a new DataFrame that contains just the Coffee House coupons

```python
#creating DataFrame only for Coffee House Coupons
data_coffee = data[(data['coupon'] == "Coffee House")]
```

```python
#looking at the acceptance rate of Coffee House coupons
data_coffee["Coupon Acceptance"].value_counts(normalize = True)
```

```python
#looking at the accceptance rates grouped by "temperature"
data_coffee.groupby("temperature")["Coupon Acceptance"].value_counts(normalize = True).unstack()
```

The above chart shows that there is slighty a higher acceptance rate if the temperature is warmer.


Next we are making a bar chart to compare the acceptance rates between those drivers who go to coffee houses less than once a month and also the direction is the same way as they are heading. 

```python
#defining variables for categories of times visited a coffee house.  less than 1 or more than 1
coffee_lt1 = ["never", "less1"]
coffee_gte1 = ["1~3", "4~8", "gt8"]


#Making a dateframe for coffee house visits that are less than 1
lt1_coupon_acceptance = data_coffee[(data_coffee["CoffeeHouse"].isin(coffee_lt1)) & (data_coffee["direction_same"] == 1)]["Coupon Acceptance"].value_counts(normalize=True).reset_index()
lt1_coupon_acceptance.columns = ['Coupon Acceptance', 'Proportion']
lt1_coupon_acceptance['Group'] = 'Less than 1 time/month'

#making a dataframe for coffee house visits that are 1 or more
gte1_coupon_acceptance = data_coffee[(data_coffee["CoffeeHouse"].isin(coffee_gte1)) & (data_coffee["direction_same"] == 1)]["Coupon Acceptance"].value_counts(normalize=True).reset_index()
gte1_coupon_acceptance.columns = ['Coupon Acceptance', 'Proportion']
gte1_coupon_acceptance['Group'] = '1 or more times/month'

# Combine both DataFrames
combined_df = pd.concat([lt1_coupon_acceptance, gte1_coupon_acceptance])

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(data=combined_df, x='Coupon Acceptance', y='Proportion', hue='Group')

# Customize the plot
plt.xlabel('Coupon Acceptance')
plt.ylabel('Proportion')
plt.title('Coupon Acceptance Proportion by Coffee House visits per month.\nCoupon is the same direction as car is heading')
plt.legend(title='Coffee House Visits')

# Show the plot
plt.show()
```

As we can see from the chart those who go to coffee house at least once are more likely to accept the coupon when the coffee house is on the way they are going.

```python
#checking to see if time plays a factor
data_coffee.groupby("time")["Coupon Acceptance"].value_counts(normalize = True).unstack()
```

Most likely time to accept a coupon is 10am

```python
#histogram graph to show how time is a factor for coupon acceptance for the 2h coupon expiration
fig = px.histogram(data_coffee.query("expiration == '2h'"), x = "time", color = "Coupon Acceptance")

fig.update_layout(
    title="Acceptance by time of day for 2h coupon expiration",
    title_font=dict(size=16),
    title_x=0.5  
)
```

```python
#histogram graph to show how time is a factor for coupon acceptance for the 1d coupon expiration
fig = px.histogram(data_coffee.query("expiration == '1d'"), x = "time", color = "Coupon Acceptance")

fig.update_layout(
    title="Acceptance by time of day for 1d coupon expiration",
    title_font=dict(size=16),
    title_x=0.5  
)
```

```python
# Convert "time" to categorical with specific order
time_order = ['7AM', '10AM', '2PM', '6PM', '10PM']
temp_coffee = data_coffee
temp_coffee['time'] = pd.Categorical(data_coffee['time'], categories=time_order, ordered=True)

# Group by "time" and expiration, then calculate acceptance rate
acceptance_rate = temp_coffee.groupby(['time', 'expiration'])['Coupon Acceptance'].mean().reset_index()

# Pivot the dataframe for easier plotting
acceptance_rate_pivot = acceptance_rate.pivot(index='time', columns='expiration', values='Coupon Acceptance')

# Plot the data
plt.figure(figsize=(10, 6))
for coupon_type in acceptance_rate_pivot.columns:
    plt.plot(acceptance_rate_pivot.index, acceptance_rate_pivot[coupon_type], marker='o', label=f'Coupon Type: {coupon_type}')

plt.xlabel('Time of Day')
plt.ylabel('Acceptance Rate')
plt.title('Acceptance Rate of Coupons by Delivery Time')
plt.legend()
plt.grid(True)
plt.show()
```
